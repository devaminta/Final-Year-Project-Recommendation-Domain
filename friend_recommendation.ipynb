```python
import pandas as pd
from  sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import OneHotEncoder
from datetime import datetime
```

```python
import os
import kagglehub
import glob

# Get the dataset directory
dataset_dir = kagglehub.dataset_download("arindamsahoo/social-media-users")

# Find the CSV file inside the dataset directory
csv_files = glob.glob(os.path.join(dataset_dir, "**", "*.csv"), recursive=True)

# Print the CSV file path
if csv_files:
    print("CSV file path:", csv_files[0])  # Print the first CSV file found
else:
    print("No CSV file found.")
```

```python
dataset = pd.read_csv('/root/.cache/kagglehub/datasets/arindamsahoo/social-media-users/versions/2/SocialMediaUsersDataset.csv')
dataset = dataset.head(10000)
```

```python
dataset
```

```python
interests = dataset['Interests'].str.get_dummies(', ')  # One-hot encode interests
interests.fillna(0, inplace=True)  # Replace NaN values with 0
```

```python
interests
```

```python
import pandas as pd
import numpy as np
from datetime import datetime

# Convert DOB to datetime format
dob = pd.to_datetime(dataset['DOB'])

# Get the current date
current_date = pd.Timestamp(datetime.now())

# Calculate age in years (approximate by days)
dataset['Age'] = (current_date - dob).dt.days // 365

dataset
```

```python
age = dataset[['Age']]
```

```python
gender = dataset[['Gender']]
gender_encoded = pd.get_dummies(gender)
```

```python
gender_encoded
```

```python
location = dataset[['City', 'Country']]
location_encoded = pd.get_dummies(location)
```

```python
location_encoded
```

```python
features = pd.concat([interests, gender_encoded, age, location_encoded], axis=1)
```

```python
similarity_matrix = cosine_similarity(features)
```

```python
user_id = 1
similar_users_indices = similarity_matrix[user_id - 1].argsort()[::-1]  # Sort indices in descending order
similar_users = similar_users_indices[1:6]
```

```python
# Print the similar users
for similar_user_index in similar_users:
    print(f"Similar User ID: {similar_user_index + 1}")
```
