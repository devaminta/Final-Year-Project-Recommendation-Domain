```python
# Combine text and tags into a single feature
posts_df['content'] = posts_df['text'] + " " + posts_df['tags']

print("Updated Posts DataFrame with Combined Content:")
print(posts_df[['post_id', 'content']])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Create TF-IDF vectors for combined content
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(posts_df['content'])

# Compute cosine similarity between posts
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Convert to a DataFrame for easier lookup
cosine_sim_df = pd.DataFrame(cosine_sim, index=posts_df['post_id'], columns=posts_df['post_id'])

print("Cosine Similarity Matrix:")
print(cosine_sim_df)
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(n_neighbors=10).fit(tfidf_matrix)

def get_recommendations(post_id, similarity_df, dataset, encoded_features, top_n=5):
    user_index = similarity_df.columns.get_loc(post_id)
    user_vector = encoded_features[[user_index]]

    ix = knn.kneighbors(user_vector, return_distance=False)

    closest = similarity_df.columns[ix.flatten()]

    return pd.DataFrame(closest, columns=['post_id']).merge(dataset).head(top_n)

```

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy

# Prepare the data for Surprise
reader = Reader(rating_scale=(0, interactions_df['interaction_score'].max()))
data = Dataset.load_from_df(interactions_df[['user', 'post_id', 'interaction_score']], reader)

# Split the data into train and test sets
trainset, testset = train_test_split(data, test_size=0.2)

# Train the SVD model
model = SVD()
model.fit(trainset)

# Test the model
predictions = model.test(testset)
print("Collaborative Filtering RMSE:", accuracy.rmse(predictions))
print("Collaborative Filtering MAE:", accuracy.mae(predictions))
```

```python
def hybrid_recommendation(user, post_id, cosine_sim_df, model, posts_df, top_n=2):
    # Content-based recommendations
    sim_scores = cosine_sim_df[post_id].sort_values(ascending=False)
    content_based_recommendations = get_recommendations(post_id, cosine_sim_df, posts_df, tfidf_matrix, top_n=5)
    content_based_recommendations = content_based_recommendations['post_id'].tolist()

    # Collaborative filtering predictions
    collaborative_predictions = []
    for post in posts_df['post_id']:
        pred = model.predict(user, post).est
        collaborative_predictions.append((post, pred))

    # Sort collaborative predictions by score
    collaborative_predictions = sorted(collaborative_predictions, key=lambda x: x[1], reverse=True)
    collaborative_recommendations = [x[0] for x in collaborative_predictions[:top_n]]
    print("here is the collaborative final result",collaborative_recommendations)

    # Combine recommendations and remove duplicates
    hybrid_recommendations = list(set(content_based_recommendations + collaborative_recommendations))[:top_n]
    return hybrid_recommendations

# Example: Recommend posts for User1 based on post_id 1
user = "User1"
post_id = 1
recommendations = hybrid_recommendation(user, post_id, cosine_sim_df, model, posts_df)
print(f"Hybrid Recommendations for {user} based on post {post_id}: {recommendations}")
```
