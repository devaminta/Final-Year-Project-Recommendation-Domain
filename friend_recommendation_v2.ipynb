```python
import pandas as pd
from  sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import OneHotEncoder
from datetime import datetime
```

```python
import os
import kagglehub
import glob

dataset_dir = kagglehub.dataset_download("arindamsahoo/social-media-users")

csv_files = glob.glob(os.path.join(dataset_dir, "**", "*.csv"), recursive=True)

if csv_files:
    print("CSV file path:", csv_files[0])
else:
    print("No CSV file found.")
```

```python
dataset = pd.read_csv('/root/.cache/kagglehub/datasets/arindamsahoo/social-media-users/versions/2/SocialMediaUsersDataset.csv')
dataset = dataset.head(10000)
```

```python
dataset
```

```python
interests = dataset['Interests'].str.get_dummies(', ')  # One-hot encode interests
interests.fillna(0, inplace=True)  # Replace NaN values with 0
```

```python
interests
```

```python
import pandas as pd
import numpy as np
from datetime import datetime

# Convert DOB to datetime format
dob = pd.to_datetime(dataset['DOB'])

# Get the current date
current_date = pd.Timestamp(datetime.now())

# Calculate age in years (approximate by days)
dataset['Age'] = (current_date - dob).dt.days // 365

dataset
```

```python
age = dataset[['Age']]
```

```python
location = dataset[['City', 'Country']]
location_encoded = pd.get_dummies(location).astype(int)
```

```python
location_encoded
```

```python
features = pd.concat([interests, age, location_encoded], axis=1)
```

```python
similarity_matrix = cosine_similarity(features)
cosine_sim_df = pd.DataFrame(similarity_matrix, index=dataset['UserID'], columns=dataset['UserID'])
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors

encoded_features = pd.concat([interests, age, location_encoded], axis=1)

knn = NearestNeighbors(n_neighbors=10).fit(encoded_features)

def get_recommendations(UserID, similarity_df, dataset, encoded_features, top_n=5):
    user_index = similarity_df.columns.get_loc(UserID)
    user_vector = encoded_features.iloc[[user_index]]

    ix = knn.kneighbors(user_vector, return_distance=False)

    closest = similarity_df.columns[ix.flatten()]

    return pd.DataFrame(closest, columns=['UserID']).merge(dataset).head(top_n)

```

```python
result = get_recommendations(1, cosine_sim_df, dataset, encoded_features, top_n=5)
```
